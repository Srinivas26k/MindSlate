```python
from unsloth import FastLanguageModel
import torch

model_name = "unsloth/gemma-3n-E2B-it-unsloth-bnb-4bit"

max_seq_length = 2048  # Choose sequence length
dtype = None  # Auto detection

# Load model and tokenizer
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=model_name,
    max_seq_length=max_seq_length,
    dtype=dtype,
    load_in_4bit=True,
)
```

```python
import json
import os

selected_files = [
    "/content/flashcards.json",
    "/content/reminders.json",
    "/content/summaries.json",
    "/content/todos.json"
]

file = []
for file_path in selected_files:
    try:
        with open(file_path, "r") as f:
            data = json.load(f)
            print(f"Successfully loaded {len(data)} items from {file_path}")
            file.extend(data)
    except FileNotFoundError:
        print(f"File not found: {file_path}")
    except json.JSONDecodeError:
        print(f"Error decoding JSON from: {file_path}")

# Optional: Print the number of items loaded
print(f"Loaded {len(file)} items from selected files.")
```

```python
from datasets import Dataset
import json

def format_prompt(example):
    # Ensure 'input' and 'output' keys exist and are not None
    input_text = example.get('input')
    output_data = example.get('output')

    if input_text is None or output_data is None:
        print(f"Skipping item due to missing 'input' or 'output': {example}")
        return None # Skip this example

    try:
        # Attempt to dump output_data to JSON string
        output_json_string = json.dumps(output_data)
    except (TypeError, ValueError) as e:
        print(f"Skipping item due to JSON encoding error for output: {output_data} - Error: {e}")
        return None # Skip this example

    return f"### Input: {input_text}\n### Output: {output_json_string}<|endoftext|>"

formatted_data = []
for item in file:
    formatted_item = format_prompt(item)
    if formatted_item is not None:
        formatted_data.append(formatted_item)

print(f"Number of items in 'file' before formatting: {len(file)}")
print(f"Number of items in 'formatted_data' after formatting: {len(formatted_data)}")

if not formatted_data:
    print("formatted_data is empty. Please check the format_prompt function and the input data.")
else:
    dataset = Dataset.from_dict({"text": formatted_data})
    print(f"Dataset created successfully with {len(dataset)} items.")
    print(f"First element of the dataset: {dataset[0]}")
```

```python
from peft import PeftModelForCausalLM

# Add LoRA adapters
if not isinstance(model, PeftModelForCausalLM):
    model = FastLanguageModel.get_peft_model(
        model,
        r=64,  # LoRA rank - higher = more capacity, more memory
        target_modules=[
            "q_proj", "k_proj", "v_proj", "o_proj",
            "gate_proj", "up_proj", "down_proj",
        ],
        lora_alpha=128,  # LoRA scaling factor (usually 2x rank)
        lora_dropout=0,  # Supports any, but = 0 is optimized
        bias="none",     # Supports any, but = "none" is optimized
        use_gradient_checkpointing=False,  # Unsloth's optimized version
        random_state=3407,
        use_rslora=False,  # Rank stabilized LoRA
        loftq_config=None, # LoftQ
    )
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments

# Training arguments optimized for Unsloth
trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,
    dataset_text_field="text",
    max_seq_length=max_seq_length,
    dataset_num_proc=2,
    args=TrainingArguments(
        per_device_train_batch_size=1, # Reduced batch size
        gradient_accumulation_steps=8,  # Increased accumulation steps (Effective batch size = 1 * 8 = 8)
        warmup_steps=10,
        num_train_epochs=3,
        learning_rate=2e-4,
        fp16=not torch.cuda.is_bf16_supported(),
        bf16=torch.cuda.is_bf16_supported(),
        logging_steps=25,
        optim="adamw_8bit",
        weight_decay=0.01,
        lr_scheduler_type="linear",
        seed=3407,
        output_dir="outputs",
        save_strategy="epoch",
        save_total_limit=2,
        dataloader_pin_memory=False,
        report_to="none",
    ),
)
```

```python
# Train the model
trainer_stats = trainer.train()
```

```python
from unsloth.chat_templates import get_chat_template
tokenizer = get_chat_template(
    tokenizer,
    chat_template = "gemma-3",
)
messages = [{
    "role": "user",
    "content": [{
        "type" : "text",
        "text" : "Continue the sequence: 1, 1, 2, 3, 5, 8,",
    }]
}]
inputs = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt = True, # Must add for generation
    return_tensors = "pt",
    tokenize = True,
    return_dict = True,
).to("cuda")
outputs = model.generate(
    **inputs,
    max_new_tokens = 64, # Increase for longer outputs!
    # Recommended Gemma-3 settings!
    temperature = 1.0, top_p = 0.95, top_k = 64,
)
tokenizer.batch_decode(outputs)
```
